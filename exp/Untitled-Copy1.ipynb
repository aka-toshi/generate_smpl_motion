```python
#骨格の長さを計算する関数
def bonelen(trc):
    joint = trc[:,1:]
    #配列作り
    a = np.arange(17)
    a[4] = 1
    a[7] = 1
    a[10] = 1
    a[13] = 0
    a[14] = 0
    a[-2:] -=1
    c = np.zeros(17*3,dtype='int32')
    c[::3] = a*3
    c[1::3] =a*3+1
    c[2::3] =a*3+2
    jointbox = (joint[:,1*3:]-joint[:,c])**2
    jointbox = np.reshape(jointbox,(len(jointbox),17,3))
    jointlen = np.sum(jointbox,axis=2)**0.5
    return jointlen
```

```python
#骨格の長さを計算する関数
#def bonelen(trc):
def bonelen(trc):
    joint = trc[:,1:]
    #配列作り
    a = np.arange(17)
    a[4] = 1
    a[7] = 1
    a[10] = 1
    a[13] = 0
    a[14] = 0
    a[-2:] -=1
    c = np.zeros(17*3,dtype='int32')
    c[::3] = a*3
    c[1::3] =a*3+1
    c[2::3] =a*3+2
    jointbox = (joint[:,1*3:]-joint[:,c])**2
    jointbox = np.reshape(jointbox,(len(jointbox),17,3))
    jointlen = np.sum(jointbox,axis=2)**0.5
    return jointlen
def jointvar(trc):
    joint = trc[:,1:]
    jmean = np.mean(joint,axis=0)
    jointbox = (joint-jmean)**2
    jointbox = np.reshape(jointbox,(len(jointbox),18,3))
    jointlen = np.sum(jointbox,axis=2)**0.5
    return np.var(jointlen,axis=0)**0.5/np.mean(jointlen,axis=0)
def jointvar1(trc):
    joint = trc[:,1:]
    nosejoint = np.zeros(np.shape(joint))
    for i in range(18):
        nosejoint[:,i*3]   = joint[:,0]  
        nosejoint[:,i*3+1] = joint[:,1]
        nosejoint[:,i*3+2] = joint[:,2]
#    nosejoint[:,::3]  += joint[:,0]
#    nosejoint[:,1::3] += joint[:,1]
#    nosejoint[:,2::3] += joint[:,2]
    joint = joint-nosejoint
    jmean = np.mean(joint,axis=0)
    jointbox = (joint-jmean)**2
    jointbox = np.reshape(jointbox,(len(jointbox),18,3))
    jointlen = np.sum(jointbox,axis=2)**0.5
    jvar =  np.var(jointlen,axis=0)**0.5
    print(jvar)
#    jvar =  np.var(jointlen,axis=0)**0.5/np.mean(jointlen,axis=0)
    lst = [4,7,10,13]
#    return np.var(jointlen,axis=0)**0.5/np.mean(jointlen,axis=0)
    return jvar[lst]
    
def footlen(trc):
    joint = trc[:,1:]
    rjoint = joint[:,30:33]
    ljoint = joint[:,39:42]
    foot = np.sum((rjoint-ljoint)**2,axis=1)**0.5
    return np.max(foot)
    
```
